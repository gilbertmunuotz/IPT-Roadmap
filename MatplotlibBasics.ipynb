```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# import matplotlib.pyplot
# version = matplotlib.__version__
# print(version)
```

### Basic Graph

```python
# Define data in Array format
x = [1,2,3]
y = [4,5,6]

plt.plot(x,y, label = "2x", color = "red")

plt.title("FIrst Graph")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# plt.xticks([2,4,6,8])
# plt.yticks([3,5,7,9])

plt.show()
```

```python
a = [6,7,8]
b = [9,10,11]

plt.plot(a,b, label = "3x", color = "Green", linewidth = 3)
plt.title("Demo Graph", fontdict={"fontsize": 15})
plt.show()
```

```python
a = [60,70,80]
b = [90,100,110]

plt.plot(a,b, color = "Purple", linewidth = 3, marker=".", markersize = 15, markeredgecolor = "yellow")
plt.title("Demo Graph", fontdict={"fontsize": 25})
plt.show()
```

```python
a = [20,30,40]
b = [50,60,70]

plt.plot(a,b, linestyle =  "--")
plt.show()
```

### Complex Graphs

```python
# Examples
x2 = np.arange(0.4,5,0.5)
x3 = np.arange(0.4,5,0.5)
plt.plot(x2, x2**2, marker = ".", markersize = 15)
plt.plot(x3, x3*2, marker = "*", markersize = 15)
plt.title("Complex Graph", fontdict = {"fontsize": 25})

# Saving Graphs
plt.savefig("cmpgraph.png", dpi = 200)

plt.show()
```

### Bar Charts

```python
labels = ["x", "Y", "Z"]
values = [20, 30 ,40]

plt.bar(labels, values)
plt.show()
```
