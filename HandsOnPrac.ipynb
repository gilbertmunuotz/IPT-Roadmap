```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Datasets via Pandas

```python
gas_data = pd.read_csv("gas_prices.csv")
print(gas_data)
```

### Displaying Values

```python
plt.title("Gas Prices in USD")
plt.plot(gas_data.Year, gas_data.Germany, label = "Germany", marker = "*", markersize = 15)
plt.plot(gas_data["Year"], gas_data["France"], label = "France", marker = ".", markersize = 15)
plt.plot(gas_data.Year, gas_data.Australia, label = "Australia", marker = "^", markersize = 15)

plt.xlabel("Year")
plt.ylabel("Gas Price")
plt.legend()

plt.savefig("Gas Price Graph.png", dpi = 500)

plt.show()
```

### FIFA Data

```python
fifa_data = pd.read_csv("fifa_data.csv")
print(fifa_data.head(3))
```

### Histogram

```python
plt.hist(fifa_data.Overall, bins = 10, color = "#F27157", edgecolor = "black")

plt.ylabel("No of Players")
plt.xlabel("Skill Level") 
plt.show()
```

### Pie Charts

```python
left_foot = fifa_data.loc[fifa_data["Preferred Foot"] == "Left"]
right_foot = fifa_data.loc[fifa_data["Preferred Foot"] == "Right"]
print(left_foot.shape[0]) # number of rows where Preferred Foot is Left
print(right_foot.shape[0]) # number of rows where Preferred Foot is Right
```

### Drawing the Pie

```python
# Grab the values first
values = [left_foot.shape[0], right_foot.shape[0]]
labels = ["Left foot", "Right foot"]
plt.pie(values, labels = labels, autopct = "%.2f%%")
plt.title("Chart Graph of Foot Preference")
plt.show()
```

```python
weight = fifa_data['Weight'].dropna().str.replace("lbs", "", regex = False).astype(int)

print(weight)
```

### Filter Data 

```python
light = weight[weight < 150].count()
medium = weight[(weight >= 150) & (weight < 180)].count()
heavy = weight[weight >= 180].count()
print(light)
print(medium)
print(heavy)
```

### Plot Data via Pie Chart

```python
values = [light, medium, heavy]
labels = ["Light", "Medium", "Heavy"]
plt.pie(values, labels = labels, autopct = "%.2f%%")
plt.title("Graph of Player Weights")
plt.show()
```

### Box and Whiskers

```python
man_utd = fifa_data.loc[fifa_data.Club == "Manchester United"]["Overall"]
chelsea = fifa_data.loc[fifa_data.Club == "Chelsea"]["Overall"]
print(man_utd)
print(chelsea)
```

### Draw a boxplot

```python
plt.figure(figsize =  (5,8))

labels = ["Man UTD", "Chelsea"]
plt.boxplot([man_utd, chelsea], tick_labels = labels)

plt.ylabel("FIFA Overall Rating")
plt.title("Football Soccer Comparison")

plt.savefig("FIFA Stats", dpi = 1000)
plt.show()
```
