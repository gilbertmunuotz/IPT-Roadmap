```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Datasets via Pandas

```python
gas_data = pd.read_csv("gas_prices.csv")
print(gas_data)
```

### Displaying Values

```python
plt.title("Gas Prices in USD")
plt.plot(gas_data.Year, gas_data.Germany, label = "Germany", marker = "*", markersize = 15)
plt.plot(gas_data["Year"], gas_data["France"], label = "France", marker = ".", markersize = 15)
plt.plot(gas_data.Year, gas_data.Australia, label = "Australia", marker = "^", markersize = 15)

plt.xlabel("Year")
plt.ylabel("Gas Price")
plt.legend()

plt.savefig("Gas Price Graph.png", dpi = 500)

plt.show()
```
